```python
def f(x):
    y_1 = str(x)
    #print(y_1)
    
    Sectornames=usadata2['Economic Sector']
    emisioninfo=usadata2[y_1]

    gasname=usadata1['Gas'][0:4]
    gasvalue=usadata1[y_1][0:4]
    fig = plt.gcf()
    ax = fig.add_subplot()
    fig.set_size_inches(16, 10)
    
    sizes=usadata2[y_1][0:5]
    label=Sectornames=usadata2['Economic Sector'][0:5]
    squarify.plot(sizes=sizes,text_kwargs={'fontsize':20}, label=label,  alpha=0.6 )
    plt.title(y_1 + ' USA Gas Emissions by Sector',fontsize=23,fontweight="bold")
    plt.axis('off')
    plt.show()



    ##########  Personalized shape(cloud )  "pie" chart     ##############3
    data = {'Carbon Dioxide':(round(gasvalue[0]/80)), 'Metehane':(round(gasvalue[1]/80)) , 'Nitrous oxide':(round(gasvalue[2]/80)), 'Fluorinated gases':(round(gasvalue[3]/80)) }
    fig = plt.figure(figsize=(20,15),
        FigureClass=Waffle, 
        title={'label': y_1 + ' USA Gas Emissions by Gas Type', 'loc': 'left','fontsize':30},
        rows=5, 
        values=data, 
        colors=("blue", "red", "green","orange"),
        labels=[f"{k} ({int(v / sum(data.values()) * 100)}%)" for k, v in data.items()],
        legend={
        # 'labels': [f"{k} ({v}%)" for k, v in data.items()],  # lebels could also be under legend instead
        'loc': 'upper left',
        'bbox_to_anchor': (1.05, 1),
        'fontsize': 25
        },             
        icons='cloud', icon_size=30, 
        icon_legend=True,

    )
year = np.arange(1990,2018,1)

x1 = widgets.IntSlider(
    value=5,
    min=np.amin(year),
    max=np.amax(year),
    step=1,
    description='Year: ',
    
    #layout=Layout(width='500px')
    )
interact(f,x=x1)


#treemap

```
